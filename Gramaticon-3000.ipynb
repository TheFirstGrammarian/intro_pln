## 2. Ejecución condicional

Retomemos los booleanos. ¿Para qué nos pueden servir? Bueno, podemos especificarle a nuestro programa que si cierta condición es verdadera, entonces que ejecute cierta acción, y que si es falsa, que se ejecute otra. Veamos un ejemplo:

```
cadena = "abcde"
if cadena.isalpha():
  print("Estamos ante una cadena hecha de caracteres alfabéticos")
else:
  print("la cadena no tiene caracteres alfabéticos")
```

Los métodos como .isalpha() no son la única manera de obtener valores booleanos. También podemos comparar dos valores, ya sean cadenas, ints, floats, etc. mediante varios operadores:

**a == b**    -> a es igual a b 

**a != b**      -> a no es igual a b

**a < b**    -> a es menor a b

**a <= b**     -> a es menor o igual a b

**a > b**       -> a es mayor que b

**a >= b**     -> a es mayor o igual a b

```
a = 3
b = 5

if a > b:
  print("a es mayor que b")
elif a < b:
  print("a es menor que b")
else:
  print("a y b son iguales")
```

Bien, ya resolvimos la primera parte del problema. Sin embargo, el Gramaticón-3000 no solo imprime un mensaje, sino que vuelve a pedirle al usuario que ingrese una respuesta adecuada.

¿Cómo hacemos esto? Aquí nos viene bien el concepto de bucle (o loop, como es conocido en inglés). Un  bucle consiste en un grupo de sentencias (líneas de código), que se van a ejecutar repetidamente hasta que se cumpla determinada condición. Los bucles se pueden formar con sentencias **while** o **for**. Veamos un primer ejemplo:

```
n = 5
while n > 0:
  print(str(n) + ' es mayor que 0')
  n = n-1
```

```
#¿qué pasa si ejecutamos este bucle cuya condición es siempre 'True'?
while True:
  print("MFNT es la mejor materia del mundo")
```

A veces, puede que no sepamos cuántas veces queremos que se itere nuestro código, es decir, no sabemos con exactitud cómo cortarlo. Para ello, podemos utilizar la sentencia **break**

```
while True:
  respuesta = input('¿Cuál es la mejor materia de Letras?: ')
  if respuesta.upper() == 'MFNT':
    print('Lo sabíamos, ¡muchas gracias! :)')
    break
  
```

# POS-tagging

Hasta aquí tenemos el armazón básico de nuestro programa. Ahora viene el meollo de la cuestión, que es lograr darle a cada palabra, de cada oración, su etiqueta morfológica. Para eso, vamos a hacer uso de la libería NLTK (natural language toolkit) que, si se dieron cuenta, instalamos al principio de esta Jupyter Notebook. NLTK nos provee varios recursos para analizar lenguaje natural, entre ellos, los POS taggers. Lástimosamente, no vamos a entrar en detalle sobre los POS tagger en Python, ni mucho menos en NLTK, pues son temas muy amplios que nos llevarían muchísimo tiempo de abordar. 

En todo caso, diremos que NLTK nos deja entrenar nuestros propios POS taggers que, mirando un corpus etiquetado, "aprenden" a etiquetar automáticamente. El trabajo ya lo tenemos hecho, y nuestro POS tagger entrenado está listo para usar:

```
NaiveBayesTagger = pickle.load(open("/content/drive/My Drive/Colab Notebooks/NaiveBayesTagger.p", "rb")) #diferencia entre r y rb
```

Nuestro tagger tiene el método .tag() asociado con él. Probémoslo con una oración y veamos qué pasa:

```
oracion = 'Monitos en la cama saltan sin parar'
```

```
NaiveBayesTagger.tag(oracion)
```

Nuestro tagger recibe como input una lista de palabras, y nos devuleve, como output, una lista de tuplas (DEFINICION) y arriba interpreta cada caracter como una palabra distinta. Python no entiende que un espacio en blanco delimita una palabra distinta, por ello, es necesario que a la oración la convirtamos en una lista de tokens, es decir, de palabras.

```
oracion_tokenizada = word_tokenize(oracion)
```

```
oracion_tokenizada
```

```
NaiveBayesTagger.tag(oracion_tokenizada)
```
